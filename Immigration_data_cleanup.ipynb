```python
immigration_df = immigration_df[immigration_df['GEO'] != 'Canada']
immigration_df
```

```python
final_df = immigration_df.groupby(['REF_DATE', 'GEO']).sum()
final_df.head()
```

```python
final_df = final_df.reset_index()
final_df.head()
```

```python
# final_df.pivot(index="REF_DATE", columns="variable", values="value")
df = pd.melt(final_df, id_vars=['REF_DATE', 'GEO'], value_vars=[
    'status_before_1980', 'status_1980_to_1990', 'status_1991_to_2000', 'status_2001_to_2010', 'status_2011_to_2015', 'status_2016_to_2021'
])
df.head()
```

```python
df['status_year'] = df['variable'].str.extract(r'_(\d{4})$')
df
```

```python
df.drop('variable', axis=1, inplace=True)
```
